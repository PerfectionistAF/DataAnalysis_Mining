```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

x, y = np.meshgrid(x, y)
eq = 0.12 * x + 0.01 * y + 1.09

fig = plt.figure()

ax = fig.gca(projection='3d')

ax.plot_surface(x, y, eq)

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D

#Import the data
data = pd.read_csv("house_price.csv")
#print(data)

X= data.iloc[:, :-1].values  
y= data.iloc[:, 2].values 
#print(X)
#print(Y)


############################# 
# Multilinear regression model
from sklearn.linear_model import LinearRegression 

reg = LinearRegression()
model = reg.fit(X, y)
print("regression coefficients: ", reg.coef_)
print("regression intercept: ", reg.intercept_)
print("model evaluation (r^2)", reg.score(X, y)) #R^2
# Predicting values
print("regression predictions: ")
print(reg.predict([[2567, 5]]))
print(reg.predict([[1200, 2]]))
print(reg.predict([[825, 5]]))
print(reg.predict([[1825, 2]]))
print(reg.predict([[1203, 3]]))



###############################################
#(3D Scatter) Plot the data and regression model

X0 = X[:, 0]
X1 = X[:, 1]

fig = plt.figure()
ax = fig.add_subplot(projection ='3d')
 
ax.scatter(X0, X1, y, color='orange') #area, rooms, price

#Plotting the regression plane
x_pred = np.linspace(min(X0), max(X0), num=30)      # range of area values
y_pred = np.linspace(min(X1), max(X1), num=30)  # range of rooms values
xx_pred, yy_pred = np.meshgrid(x_pred, y_pred)
model_viz = np.array([xx_pred.flatten(), yy_pred.flatten()]).T
predicted_plot = model.predict(model_viz) #predicted z values
ax.scatter(xx_pred.flatten(), yy_pred.flatten(), predicted_plot, facecolor=(0,0,0,0), s=10, edgecolor='#70b3f0')

#ax.view_init(45, 0)

ax.set_xlabel('Area')
ax.set_ylabel('Rooms')
ax.set_zlabel('Price')

fig.suptitle("Multilinear Regression Model")
plt.show()



########################
# Detecting and plotting outliers

predicted = model.predict(X)
residuals = list(np.subtract(y, predicted))
standardized_residuals = residuals / np.std(residuals) # |value| > 2 is outlier
print("\n", y, "\n")
print(predicted, "\n")
print(standardized_residuals, "\n")

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(X[:, 0], X[:, 1], standardized_residuals)
ax.set_xlabel('Area')
ax.set_ylabel('Rooms')
ax.set_zlabel('Residuals')
fig.suptitle("Standardized Residuals")
plt.show()
```

```python
import pywt
 
x = np.linspace(0, 1, num=2048)
chirp_signal = np.sin(250 * np.pi * x**2)
    
fig, ax = plt.subplots(figsize=(6,1))
ax.set_title("Original Chirp Signal: ")
ax.plot(chirp_signal)
plt.show()
    
data = chirp_signal
waveletname = 'sym5'
 
fig, axarr = plt.subplots(nrows=5, ncols=2, figsize=(6,6))
for ii in range(5):
    (data, coeff_d) = pywt.dwt(data, waveletname)
    axarr[ii, 0].plot(data, 'r')
    axarr[ii, 1].plot(coeff_d, 'g')
    axarr[ii, 0].set_ylabel("Level {}".format(ii + 1), fontsize=14, rotation=90)
    axarr[ii, 0].set_yticklabels([])
    if ii == 0:
        axarr[ii, 0].set_title("Approximation coefficients", fontsize=14)
        axarr[ii, 1].set_title("Detail coefficients", fontsize=14)
    axarr[ii, 1].set_yticklabels([])
plt.tight_layout()
plt.show()
```
