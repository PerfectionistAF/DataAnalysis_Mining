```python
import numpy as np
import pandas as pd
```

```python
#read dataset as .csv file with \t as a delimiter
DataFrame = pd.read_csv("weather_nominal.csv", delimiter = ',', header = 0, index_col = 0)
#make sure all columns are of type str
DataFrame = DataFrame.astype({'windy':'string'})
print(DataFrame)
```

# Find Frequent Itemsets Using Apriori 

```python
from apyori import apriori
```

```python
#shape of dataset (14, 6)
#association rules from transactions
tuples = []
for i in range(0, 14):
    tuple = []
    for j in range(0, 5):
        val = DataFrame.values[i, j]
        if str(val) != 'nan':
            tuple.append(val)
    tuples.append(tuple)
    tupleslist = DataFrame.values.tolist()
    print(tupleslist[i],'\n')    
```

```python
mylist = [[x for x in sublist if str(x)!= 'nan']
         for sublist in tupleslist]
for a in range(len(mylist)):
    print(mylist[a],'\n')
```

# Apply Apriori With Required Threshold 

```python
#minsup = 10%, minconf = 70%
Association_Rules = apriori(tuples, min_support = 0.1, min_confidence = 0.7,
                            min_length = 1)
Association_Results = list(Association_Rules)
print("{} derived relations".format(len(Association_Results)))
#print(Association_Results)
```

```python
#Convert into rules
for item in Association_Results:
    print(item)
    pair = item[0]
    items = [y for y in pair]
    print("Frequent Itemsets" + str(items), '\n')
    print("Support" + str(item[1]), '\n')
    if(len(pair) > 1):
        for rule in item[2]:
            print("Rule: " + str(rule[0]) + "->" + str(rule[1]) , "\n")
            print("Confidence: " + str(rule[2]), "\n")
            print("Lift: " + str(rule[3]), "\n")
            print("_______________________________________________________________________")
```
